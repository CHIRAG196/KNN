```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split as tts, GridSearchCV
from sklearn.metrics import confusion_matrix, r2_score, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("breast_cancer.csv")
data.head()
# data.describe()
```

```python
print(data.columns)
```

```python
print(data.info())
```

```python
data['diagnosis'] = data['diagnosis'].replace({'M':1,"B":0})
data.head()
```

```python
data.drop(['id','Unnamed: 32'],1,inplace = True)
print(data.head())
```

```python
def no_zeros():
    for column in data:
        data[column] = data[column].replace(0, NaN)        # replace_with_NaN
        mean = int(data[column].mean(skipna=True))
        data[column] = data[column].replace(NaN, mean)                 #replace_with_mean
```

```python
data.head()
print(data.info())
```

```python
data.head()
```

```python
#split the data in training and test data
X = data.iloc[:,12:22]
y = data.iloc[0:,0]
X_train, X_test, y_train, y_test = tts(X,y,random_state = 42, test_size=0.3)
print(data.shape[0])
print(X_train.shape[0])
print(X_test.shape[0])
```

```python
#Scaling the data
#The data is first splitted and than scaled.
std_scaler = StandardScaler()
X_train_std_scaler = std_scaler.fit_transform(X_train)  # first fit to data and than transform it
X_test_std_scaler = std_scaler.transform(X_test)        # performs standardization by centering and scaling
                                                       ## Centering is a form of preprocessing only numerical data in which
                                                      ### mean of variable is subtracted with every other variable.   AND 
                                                     #### Scaling is the same but just multiplying rather than subtracting.        
```

```python
#finding value for k

import math 

K = len(data)
K = math.sqrt(K)
K 
                         #if K=even than subtract 1 for final K value.             
```

```python
#Initializing the module
classifier = KNeighborsClassifier(n_neighbors = 23, metric = 'euclidean')
classifier
```

```python
classifier.fit(X_train, y_train)
y_pred = classifier.score(X_test,y_test)             #this is y_pred
y_pred
```

```python
# part of GridSearchCV
parameter = {"n_neighbors":np.arange(1,10,1), "metric":["euclidean", "minkowski", "jaccard", "cosine"]}

Grid_search_Cv = GridSearchCV(estimator = classifier, param_grid = parameter, cv = 10)
Grid_search_Cv.fit(X,y)
```

```python
accuracy = Grid_search_Cv.best_score_
accuracy
```
